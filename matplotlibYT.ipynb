```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10, 20, 15, 25]

plt.plot(x,y)
plt.show()
```

```python
a = ['mon','tues','wed','thus','fri']
b= [10,15,7,20,12]
plt.plot(a,b)
plt.title('bakery sales this week')
plt.xlabel('day of week')
plt.ylabel('sales per day')
plt.show()
```
