```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10, 20, 15, 25]

plt.plot(x,y)
plt.show()
```

```python
a = ['mon','tues','wed','thus','fri']
b= [10,15,7,20,12]
plt.plot(a,b)
plt.title('bakery sales this week')
plt.xlabel('day of week')
plt.ylabel('sales per day')
plt.show()
```

```python
months = [1,2,3,4]
sales = [1600,1200,1800,1100]
plt.plot(months, sales, color = 'green', 
         linestyle = '-.', linewidth=2, marker = 'd', 
         label='2015 sales data')
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("monthly sales report")
plt.legend(loc='upper left', fontsize=10)
plt.grid(color="gray", linestyle=":", linewidth=1)
plt.xlim(1,4)
plt.ylim(1000,2000)
plt.xticks([1,2,3,4], ['M1','M2','M3','M4'])
plt.show()
```

```python
#barchart
import matplotlib.pyplot as plt

product = ['A',"B","C","D"]
sales = [1000,1500,800,1200]

plt.barh(product,sales,color='crimson',label='sales 2015')
plt.xlabel('product')
plt.ylabel('sales')
plt.title("product sale comparison")
plt.legend()
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt
#autopct='$1.1%', '$1.1%f'
 
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', 
        colors= ['cyan','turquoise','aquamarine', 'teal'])
plt.title('Revenue Contribute by Region')
```

```python
#histogram
import matplotlib.pyplot as plt

scores= [60,55,67,85,81,99,73,79,64,83,77,50,59,61,66,95,91,93,87,74]

plt.hist(scores, bins=5, color='salmon', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Distrubution Of Student')
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt

hours_study = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,80,90,95]

plt.scatter(hours_study, exam_scores, 
            color='red', marker='s',
           label= 'Student Data')
plt.xlabel('Hours studied')
plt.ylabel('Exam score')
plt.title('Relationship between study time and exam score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt

plt.scatter([1,2,3],[50,55,60], color='blue', label='Class A')
plt.scatter([1,2,3],[45,50,52], color='orange', label='Class B')

plt.xlabel('Hours studied')
plt.ylabel('Exam score')
plt.title('Relationship between two class')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,30,20,40]

plt.subplot(1,2,1) # 1 row, 2 column, 1st subplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2) # 1 row, 2 column, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2, figsize=(10,5))

x = [1,2,3,4]
y = [10,30,20,40]

ax[0].plot(x,y, color='blue')
ax[0].set_title('Line Plot')

ax[1].bar(x,y, color= 'green')
ax[1].set_title('Bar chart')

fig.suptitle('comparison of line and bar charts')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,35]

plt.plot(x,y, color='magenta', marker = 'o')
plt.title('simple line plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.savefig('line_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
